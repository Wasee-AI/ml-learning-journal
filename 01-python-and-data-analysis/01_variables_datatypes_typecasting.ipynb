## Part 1 — Variables

A **variable** is a named container that stores a value in memory. Python does not require you to declare the type — it infers it automatically.

In a stock analysis, every data point I work with — price, market cap, PE ratio, company name — is stored as a variable first.

```python
# ── STOCK SNAPSHOT: Skipper Ltd (as of analysis date) ──────────────────────

company_name   = "Skipper Ltd"        # str  — company identifier
ticker         = "SKIPPER"            # str  — NSE ticker symbol
sector         = "Capital Goods"      # str  — sector classification

current_price  = 380.0                # float — Current Market Price (CMP) in ₹
market_cap     = 4296                 # int   — Market cap in ₹ Crore
outstanding_shares = 11.29            # float — Shares outstanding in Crore

current_pe     = 23.4                 # float — current trailing P/E
industry_pe    = 44.4                 # float — sector / industry P/E
ten_yr_avg_pe  = 20.9                 # float — 10-year historical average P/E

is_under_radar = True                 # bool  — my personal flag: tracking closely?

print(f"Company  : {company_name} ({ticker})")
print(f"Sector   : {sector}")
print(f"CMP      : ₹{current_price}")
print(f"Mkt Cap  : ₹{market_cap} Cr")
print(f"P/E      : {current_pe} | Industry P/E: {industry_pe} | 10yr Avg: {ten_yr_avg_pe}")
print(f"On radar : {is_under_radar}")
```

**Observation:** Notice that `current_price`, `current_pe`, `industry_pe` are all `float` — they carry decimal precision which matters in financial calculations. `market_cap` is `int` here since we're working in whole crores. This distinction matters when we typecast later.

## Part 2 — Data Types

Python has 7 core data types. Each one has a specific use in a stock analysis workflow. Below I map each type to where it naturally appears in my analysis.

```python
# ── INT — whole numbers ─────────────────────────────────────────────────────
# Use case: quantities, crore values, share counts, years

market_cap_cr       = 4296          # ₹ Crore (whole number sufficient)
shares_outstanding  = 112893293     # exact share count (from exchange filing)
analysis_year       = 2026
target_horizon_days = 365           # 12-month investment horizon

print("── INT examples ──")
print(f"Market Cap : ₹{market_cap_cr} Cr")
print(f"Shares     : {shares_outstanding:,}")
print(f"Horizon    : {target_horizon_days} days")
print(f"Type check : {type(market_cap_cr)}")
```

```python
# ── FLOAT — decimal numbers ─────────────────────────────────────────────────
# Use case: prices, ratios, margins, percentages — precision matters here

cmp             = 380.0     # Current Market Price
eps_ttm         = 16.24     # Earnings Per Share (trailing twelve months)
opm_q3fy26      = 11.8      # Operating Profit Margin % for Q3 FY26
debt_to_equity  = 0.43      # D/E ratio (lower is safer)
roe             = 18.6      # Return on Equity %

print("── FLOAT examples ──")
print(f"CMP        : ₹{cmp}")
print(f"EPS (TTM)  : ₹{eps_ttm}")
print(f"OPM%       : {opm_q3fy26}%")
print(f"D/E Ratio  : {debt_to_equity}")
print(f"ROE        : {roe}%")
print(f"Type check : {type(cmp)}")
```

## Part 3 — Typecasting

**Typecasting** means converting a value from one data type to another. In stock analysis, this happens constantly — especially when reading data from CSV files, APIs, or copy-pasting from screeners, where numbers often arrive as strings.

### Why this matters in practice:
When you download stock data from a screener website or read a CSV, numeric values often come in as `str`. If you try to do arithmetic on a string, Python throws an error. Typecasting fixes this.

```python
# ── REAL SCENARIO: Data pasted from a screener or CSV ──────────────────────
# Screener.in / Tickertape exports sometimes give numbers as strings

raw_cmp         = "380.50"     # came in as string from a data export
raw_market_cap  = "4296"       # string
raw_pe          = "23.4"       # string
raw_eps         = "16.24"      # string

print("Before typecasting:")
print(f"  raw_cmp type        : {type(raw_cmp)}")
print(f"  raw_market_cap type : {type(raw_market_cap)}")

# If I try to calculate without casting, Python will fail
# print(raw_cmp + raw_pe)  # This would give '380.5023.4' — string concatenation, NOT addition!
```

```python
# ── CASTING to correct types ────────────────────────────────────────────────

cmp         = float(raw_cmp)        # str → float
market_cap  = int(raw_market_cap)   # str → int
pe          = float(raw_pe)         # str → float
eps         = float(raw_eps)        # str → float

print("After typecasting:")
print(f"  cmp        : {cmp}   → type: {type(cmp)}")
print(f"  market_cap : {market_cap}  → type: {type(market_cap)}")
print(f"  pe         : {pe}    → type: {type(pe)}")
print(f"  eps        : {eps}  → type: {type(eps)}")

# Now arithmetic works correctly
implied_price = pe * eps
print(f"\nImplied price (P/E × EPS) : ₹{implied_price:.2f}")
print(f"  → Roughly validates CMP of ₹{cmp} ✓" if abs(implied_price - cmp) < 50 else "  → Gap detected — check data")
```

```python
# ── INT ↔ FLOAT: When does the distinction matter? ─────────────────────────

shares_cr   = 11.29             # float: shares in Crore
pat_cr      = 183.4             # float: annual PAT in ₹ Crore

# EPS calculation — must stay float for precision
eps_calc    = pat_cr / shares_cr
print(f"Calculated EPS : ₹{eps_calc:.2f}")     # float division — precise

# int() truncates — dangerous in financial calculations!
eps_truncated = int(eps_calc)
print(f"After int()    : ₹{eps_truncated}      ← truncated, NOT rounded — loses ₹{eps_calc - eps_truncated:.2f}")

# Correct rounding when you need a whole number
eps_rounded = round(eps_calc, 2)
print(f"After round()  : ₹{eps_rounded}    ← correct approach")
```

```python
# ── FLOAT → STR: Formatting output for reports ─────────────────────────────
# When building analysis reports or print summaries, convert numbers to
# formatted strings for readability

target_price    = 520.0
upside_pct      = ((target_price - cmp) / cmp) * 100

# Float to string for a clean report line
upside_label    = str(round(upside_pct, 1)) + "%"
target_label    = "₹" + str(target_price)

print("── Analyst Report Output ──")
print(f"  Stock          : Skipper Ltd (SKIPPER)")
print(f"  Recommendation : BUY")
print(f"  CMP            : ₹{cmp}")
print(f"  12M Target     : {target_label}")
print(f"  Upside         : {upside_label}")
```

```python
# ── TYPECAST ERROR — what to watch for ────────────────────────────────────
# This is a common error when loading messy data

bad_value = "N/A"   # screener sometimes returns this for missing data

try:
    converted = float(bad_value)
except ValueError as e:
    print(f"ValueError caught: {e}")
    print("  → Handle missing data before typecasting")
    print("  → In Pandas (next notebook), we use pd.to_numeric(errors='coerce') for this")

# Safe conversion with a fallback value
def safe_float(value, fallback=0.0):
    """Convert to float safely. Returns fallback if conversion fails."""
    try:
        return float(value)
    except (ValueError, TypeError):
        return fallback

print(f"\nsafe_float('N/A')    : {safe_float('N/A')}")
print(f"safe_float('23.4')   : {safe_float('23.4')}")
print(f"safe_float(None)     : {safe_float(None)}")
```

## Part 4 — `type()` and Inspection

Use `type()` to verify what Python thinks a variable is — essential when debugging data loading issues.

```python
# ── Complete type audit of a stock data record ─────────────────────────────
# This is the kind of check you run after loading data from an external source

stock_record = {
    "ticker"      : "SKIPPER",
    "cmp"         : 380.0,
    "market_cap"  : 4296,
    "pe"          : 23.4,
    "on_watchlist": True,
    "price_history": [312, 340, 378, 365, 380],
    "52wk_range"  : (198.0, 497.9)
}

print("Type audit for stock_record:")
print(f"  {'Field':<18} {'Value':<20} {'Type'}")
print("  " + "-" * 55)
for field, value in stock_record.items():
    print(f"  {field:<18} {str(value):<20} {type(value).__name__}")
```

## Summary

| Data Type | Financial Use Case | Watch Out For |
|---|---|---|
| `int` | Market cap (Cr), share count, years | Don't use for prices — loses decimal precision |
| `float` | Prices, ratios, margins, EPS | `int()` truncates, not rounds |
| `str` | Tickers, labels, recommendations | Numeric strings can't do arithmetic until cast |
| `bool` | Screening flags, signal alerts | Result of comparisons (e.g., `pe < industry_pe`) |
| `list` | Price history, quarterly series, watchlists | Ordered and changeable — good for time-series |
| `tuple` | 52wk range, fixed bands, FY labels | Immutable — use for reference data that must not change |
| `dict` | Full stock profile, quarterly record | The foundation of a Pandas DataFrame row |

### Key Typecasting Rules for Financial Data
1. **Always cast after loading** from CSV / screener exports — values often arrive as strings
2. **Never use `int()` on prices** — truncation causes silent calculation errors
3. **Use `round()`** instead of `int()` when a whole number is genuinely needed
4. **Wrap casts in `try/except`** when data quality is uncertain (`"N/A"`, `None`, empty cells)
5. **Use `type()`** to audit variables after loading any external data
